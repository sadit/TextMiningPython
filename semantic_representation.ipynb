```python
import fasttext
```

```python
import fasttext.util
fasttext.util.download_model('es', if_exists='ignore') 
ft = fasttext.load_model('cc.es.300.bin')
```

```python
ft.get_word_vector('casa')
```

```python
import numpy as np

def norm(u):
    return np.linalg.norm(u)

def dist(u, v):
    return 1.0 - np.dot(u, v) / (norm(u) * norm(v))
```

```python
w1 = ft.get_word_vector("dulce")
w2 = ft.get_word_vector("caramelo")

print((norm(w1), norm(w2)))
dist(w1, w2)
```

```python
w1 = ft.get_sentence_vector("azul")
w2 = ft.get_sentence_vector("bicicleta es verde")

print((norm(w1), norm(w2)))
dist(w1, w2)
```
